# 2.1 Dropping Columns

Drop D_87, D_88, D_110, D_111, B_39, D_73, B_42, D_134, B_29, D_132, D_76, D_142, D_53, D_50, D_105, D_56 as they have been droped in trainning set.

# 2.2 Imputing Missing Values

The operations done on the columns are same as that of the operations done in the training set

# 2.2.1 Spend Variables

```python
test["S_7"].fillna(value=test["S_7"].mean(),inplace=True)
test["S_24"].fillna(value=test["S_24"].mean(),inplace=True)
test["S_9"].fillna(value=test["S_9"].mean(),inplace=True)
test["S_25"].fillna(value=test["S_25"].mean(),inplace=True)
test["S_26"].fillna(value=test["S_26"].mean(),inplace=True)
test["S_23"].fillna(value=test["S_23"].mean(),inplace=True)
```

# 2.2.2 Risk Variables

```python
test["R_7"].value_counts()
```

R_7 has maximum values as 0. So we need to scale the larger values down.

```python
test["R_7"]=test["R_7"]/test["R_7"].max()
```

```python
test["R_14"].value_counts()
```

R_14 has maximum values as 0. So we need to scale the larger values down.

```python
test["R_14"]=test["R_14"]/test["R_14"].max()
```

Now we need to fill the null values.

```python
test["R_7"].fillna(value=test["R_7"].median(),inplace=True)
test["R_7"].fillna(value=test["R_12"].mean(),inplace=True)
test["R_7"].fillna(value=test["R_14"].median(),inplace=True)
test["R_7"].fillna(value=test["R_27"].mean(),inplace=True)
```

# 4.2.3 Balance Variables

```python
test.loc[test["B_37"].isnull() & ~test["B_25"].isnull(),"B_37"]=(test["B_37"].corr(test["B_25"])*test["B_37"].std()/test["B_25"].std())*(test.loc[test["B_37"].isnull() & ~test["B_25"].isnull(),"B_25"]-test["B_25"].mean())+test["B_37"].mean()
test.loc[~test["B_37"].isnull() & test["B_25"].isnull(),"B_25"]=(test["B_37"].corr(test["B_25"])*test["B_25"].std()/test["B_37"].std())*(test.loc[~test["B_37"].isnull() & test["B_25"].isnull(),"B_37"]-test["B_37"].mean())+test["B_25"].mean()
```

```python
test["B_37"].fillna(test["B_37"].mean(),inplace=True)
test["B_25"].fillna(test["B_25"].mean(),inplace=True)
```

```python
test.loc[test["B_13"].isnull() & ~test["B_15"].isnull(),"B_13"]=(test["B_13"].corr(test["B_15"])*test["B_13"].std()/test["B_15"].std())*(test.loc[test["B_13"].isnull() & ~test["B_15"].isnull(),"B_15"]-test["B_15"].mean())+test["B_13"].mean()
test.loc[~test["B_13"].isnull() & test["B_15"].isnull(),"B_15"]=(test["B_13"].corr(test["B_15"])*test["B_15"].std()/test["B_13"].std())*(test.loc[~test["B_13"].isnull() & test["B_15"].isnull(),"B_13"]-test["B_13"].mean())+test["B_15"].mean()
```

Fill the remaning null values.

```python
test["B_13"].fillna(test["B_13"].mean(),inplace=True)
test["B_15"].fillna(test["B_15"].mean(),inplace=True)
```

Thus B_13 and B_15 is handled.

Calculating Values of B_2 and B_3 from B_37

```python
test.loc[test["B_2"].isnull() & ~test["B_37"].isnull(),"B_2"]=(test["B_2"].corr(test["B_37"])*test["B_2"].std()/test["B_37"].std())*(test.loc[test["B_2"].isnull() & ~test["B_37"].isnull(),"B_37"]-test["B_37"].mean())+test["B_2"].mean()
test.loc[test["B_3"].isnull() & ~test["B_37"].isnull(),"B_3"]=(test["B_3"].corr(test["B_37"])*test["B_3"].std()/test["B_37"].std())*(test.loc[test["B_3"].isnull() & ~test["B_37"].isnull(),"B_37"]-test["B_37"].mean())+test["B_3"].mean()
```

Calculating B_17 from B_2

```python
test.loc[test["B_17"].isnull() & ~test["B_2"].isnull(),"B_17"]=(test["B_17"].corr(test["B_2"])*test["B_17"].std()/test["B_2"].std())*(test.loc[test["B_17"].isnull() & ~test["B_2"].isnull(),"B_2"]-test["B_2"].mean())+test["B_17"].mean()
```

# 2.2.4 Payment Variables

```python
test["P_2"].fillna(value=test["P_2"].mean(),inplace=True)
```

P_2 variable is handled

```python
test["P_3"].fillna(value=test["P_3"].mean(),inplace=True)
```

# 2.2.5 Delinquency Variables

```python
test["D_46"].fillna(value=test["D_46"].mean(),inplace=True)
test["D_62"].fillna(value=test["D_62"].mean(),inplace=True)
```

```python
test.loc[test["D_48"].isnull() & ~test["P_2"].isnull(),"D_48"]=(test["D_48"].corr(test["P_2"])*test["D_48"].std()/test["P_2"].std())*(test.loc[test["D_48"].isnull() & ~test["P_2"].isnull(),"P_2"]-test["P_2"].mean())+test["D_48"].mean()
```

```python
test["D_69"].fillna(value=test["D_69"].mean(),inplace=True)
test["D_55"].fillna(value=test["D_55"].mean(),inplace=True)
test["D_118"].fillna(value=test["D_118"].mean(),inplace=True)
test["D_121"].fillna(value=test["D_121"].mean(),inplace=True)
test["D_115"].fillna(value=test["D_115"].mean(),inplace=True)
test["D_119"].fillna(value=test["D_119"].mean(),inplace=True)
test["D_131"].fillna(value=0,inplace=True)
test["D_130"].fillna(value=0,inplace=True)
test["D_104"].fillna(value=0,inplace=True)
test["D_141"].fillna(value=0,inplace=True)
test["D_128"].fillna(value=0,inplace=True)
test["D_133"].fillna(value=test["D_133"].mean(),inplace=True)
test["D_144"].fillna(value=test["D_144"].mean(),inplace=True)
test["D_102"].fillna(value=test["D_102"].mean(),inplace=True)
test["D_52"].fillna(value=test["D_52"].mean(),inplace=True)
test["D_112"].fillna(value=1,inplace=True)
test["D_45"].fillna(value=test["D_45"].mean(),inplace=True)
test["D_54"].fillna(value=1,inplace=True)
test["D_41"].fillna(value=0,inplace=True)
```
